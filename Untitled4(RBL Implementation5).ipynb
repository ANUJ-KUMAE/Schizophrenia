```python
import numpy as np
import matplotlib.pyplot as plt
import pyedflib
```

```python
f = pyedflib.EdfReader('h01.edf')
n = f.signals_in_file
signal_labels = f.getSignalLabels()
sigbufs = np.zeros((n, f.getNSamples()[0]))
for i in np.arange(n):
    sigbufs[i, :] = f.readSignal(i)
fs = f.getSampleFrequency(0)
f.close()
```

```python
t = np.arange(0, sigbufs.shape[1]/fs, 1/fs)
fig, ax = plt.subplots(n, 1, figsize=(12, 6), sharex=True)
for i in np.arange(n):
    ax[i].plot(t, sigbufs[i, :])
    ax[i].set_ylabel(signal_labels[i])
ax[-1].set_xlabel('Time (s)')
plt.show()
```

```python
from scipy import signal

f, Pxx = signal.periodogram(sigbufs, fs, axis=1)
fig, ax = plt.subplots(n, 1, figsize=(12, 6), sharex=True)
for i in np.arange(n):
    ax[i].semilogy(f, Pxx[i, :])
    ax[i].set_ylabel(signal_labels[i])
ax[-1].set_xlabel('Frequency (Hz)')
plt.show()
```

```python
f, t, Sxx = signal.spectrogram(sigbufs, fs)
fig, ax = plt.subplots(n, 1, figsize=(12, 6), sharex=True)
for i in np.arange(n):
    im = ax[i].pcolormesh(t, f, Sxx[i, :, :], cmap='jet', shading='auto')
    ax[i].set_ylabel(signal_labels[i])
fig.colorbar(im, ax=ax.ravel().tolist())
ax[-1].set_xlabel('Time (s)')
plt.show()
```

```python
import mne
from sklearn.preprocessing import LabelEncoder
```

```python
raw_h = mne.io.read_raw_edf("h01.edf", preload=True)
raw_s = mne.io.read_raw_edf("s01.edf", preload=True)
```

```python
h01_data = raw_h.get_data().T
s01_data = raw_s.get_data().T
data = np.vstack((h01_data, s01_data))
labels = np.hstack((np.zeros(h01_data.shape[0]), np.ones(s01_data.shape[0])))
```

```python
le = LabelEncoder()
labels = le.fit_transform(labels)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.8, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
Decision_accuracy = clf.score(X_test, y_test)
print("Accuracy:", Decision_accuracy)
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(kernel = 'poly')
```

```python
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: ',accuracy)
```
