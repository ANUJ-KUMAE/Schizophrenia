```python
import numpy as np
import pandas as pd
from glob import glob
import os
import matplotlib.pyplot as plt
```

```python
import mne
```

```python
plt.style.use('seaborn-bright') # Plot Style
plt.rcParams['figure.dpi'] = 200 # DPI setting
plt.rcParams['figure.figsize'] = (4, 2)
```

```python
raw_h11 = mne.io.read_raw_edf("h11.edf", preload=True)
raw = raw_h11.pick_types(meg=False, eeg=True, eog=False, exclude='bads')
```

```python
print(raw.info)
```

```python
raw.crop(0, 60) # croping the first minute data and remove the rest
```

```python
raw.plot()
```

```python
raw.plot_psd()
```

```python
raw.resample(600) #resamples to a rate of 600 Hz
```

```python
%%capture
raw.filter(1., None)
```

```python
%%capture
raw.filter(None, 50.)
```

```python
%%capture
raw.notch_filter(60)
```

```python
# the first 60 is start (inclusive), 241 is stop (exlusive), and 60 is step
raw.notch_filter(np.arange(60, 241, 60)) 
```
